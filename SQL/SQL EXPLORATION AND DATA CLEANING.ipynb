### **CLEAN THE DATA**

```sql
--correct data types
CREATE OR REPLACE TABLE insurance_data_clean AS
SELECT 
  Customer_ID,
  Age,
  Gender,
  Province,
  TRY_CAST(NULLIF(REGEXP_REPLACE(Monthly_Income, '[R,\\s]', ''), '') AS DOUBLE) AS Monthly_Income,
  COALESCE(
    TRY_TO_DATE(Join_Date, 'yyyy-MM-dd'),
    TRY_TO_DATE(Join_Date, 'dd-MMM-yyyy'),
    TRY_TO_DATE(Join_Date, 'MM/dd/yyyy'),
    TRY_TO_DATE(Join_Date, 'dd/MM/yyyy')
  ) AS Join_Date,
  Policy_ID,
  Policy_Type,
  TRY_CAST(NULLIF(REGEXP_REPLACE(Premium_Amount, '[R,\\s]', ''), '') AS DOUBLE) AS Premium_Amount,
  Policy_Status,
  Claim_ID,
  COALESCE(
    TRY_TO_DATE(Claim_Date, 'yyyy-MM-dd'),
    TRY_TO_DATE(Claim_Date, 'dd-MMM-yyyy'),
    TRY_TO_DATE(Claim_Date, 'MM/dd/yyyy'),
    TRY_TO_DATE(Claim_Date, 'dd/MM/yyyy')
  ) AS Claim_Date,
  TRY_CAST(NULLIF(REGEXP_REPLACE(Claim_Amount, '[R,\\s]', ''), '') AS DOUBLE) AS Claim_Amount,
  Claim_Status,
  Fraud_Flag

FROM workspace.dataforge_work.insurance_data
```

```sql
--Remove duplicates
SELECT DISTINCT *
FROM insurance_data_clean
```

```sql

--Handle NULL values (COALESCE)
CREATE OR REPLACE TABLE insurance_data_clean AS
SELECT
  COALESCE(NULLIF(TRIM(Customer_ID), ''), 'Unknown') AS Customer_ID,
  COALESCE(Age, 0) AS Age,
  COALESCE(NULLIF(TRIM(Gender), ''), 'Unknown') AS Gender,
  COALESCE(NULLIF(TRIM(Province), ''), 'Unknown') AS Province,

  -- Numeric cleaning
  COALESCE(
    TRY_CAST(NULLIF(REGEXP_REPLACE(Monthly_Income, '[R,\\s]', ''), '') AS DOUBLE),
    0
  ) AS Monthly_Income,

  -- Join_Date with garbage handling
  CASE 
    WHEN Join_Date IS NULL OR TRIM(Join_Date) IN ('', 'N/A', 'UNKNOWN', '--') 
      THEN DATE('1900-01-01')
    ELSE COALESCE(
      TRY_TO_DATE(Join_Date, 'yyyy-MM-dd'),
      TRY_TO_DATE(Join_Date, 'dd-MMM-yyyy'),
      TRY_TO_DATE(Join_Date, 'MM/dd/yyyy'),
      TRY_TO_DATE(Join_Date, 'dd/MM/yyyy'),
      DATE('1900-01-01')
    )
  END AS Join_Date,

  COALESCE(NULLIF(TRIM(Policy_ID), ''), 'Unknown') AS Policy_ID,
  COALESCE(NULLIF(TRIM(Policy_Type), ''), 'Unknown') AS Policy_Type,

  COALESCE(
    TRY_CAST(NULLIF(REGEXP_REPLACE(Premium_Amount, '[R,\\s]', ''), '') AS DOUBLE),
    0
  ) AS Premium_Amount,

  COALESCE(NULLIF(TRIM(Policy_Status), ''), 'UNKNOWN') AS Policy_Status,

  -- Claim_ID with garbage handling
  COALESCE(NULLIF(TRIM(Claim_ID), ''), 'Unknown') AS Claim_ID,

  -- Claim_Date with garbage handling
  CASE 
    WHEN Claim_Date IS NULL OR TRIM(Claim_Date) IN ('', 'N/A', 'UNKNOWN', '--') 
      THEN DATE('1900-01-01')
    ELSE COALESCE(
      TRY_TO_DATE(Claim_Date, 'yyyy-MM-dd'),
      TRY_TO_DATE(Claim_Date, 'dd-MMM-yyyy'),
      TRY_TO_DATE(Claim_Date, 'MM/dd/yyyy'),
      TRY_TO_DATE(Claim_Date, 'dd/MM/yyyy'),
      DATE('1900-01-01')
    )
  END AS Claim_Date,

  COALESCE(
    TRY_CAST(NULLIF(REGEXP_REPLACE(Claim_Amount, '[R,\\s]', ''), '') AS DOUBLE),
    0
  ) AS Claim_Amount,

  COALESCE(NULLIF(TRIM(Claim_Status), ''), 'UNKNOWN') AS Claim_Status,
  COALESCE(NULLIF(TRIM(Fraud_Flag), ''), 'Unknown') AS Fraud_Flag
FROM insurance_data_clean;

SELECT *
FROM insurance_data_clean
```

```sql
--Fix inconsistent text
CREATE OR REPLACE TABLE insurance_data_clean AS
SELECT
  Customer_ID,
  Age,

  -- Gender normalization
  CASE 
    WHEN UPPER(TRIM(Gender)) IN ('M','MALE') THEN 'Male'
    WHEN UPPER(TRIM(Gender)) IN ('F','FEMALE') THEN 'Female'
    ELSE 'Unknown'
  END AS Gender,

  -- Province normalization
  CASE 
    WHEN UPPER(TRIM(Province)) IN ('WC','WESTERN CAPE') THEN 'Western Cape'
    WHEN UPPER(TRIM(Province)) IN ('EC','EASTERN CAPE') THEN 'Eastern Cape'
    WHEN UPPER(TRIM(Province)) IN ('KZN','KWAZULU-NATAL','KWAZULU NATAL') THEN 'KwaZulu-Natal'
    WHEN UPPER(TRIM(Province)) IN ('FS','FREE STATE') THEN 'Free State'
    WHEN UPPER(TRIM(Province)) IN ('GP','GAUTENG') THEN 'Gauteng'
    WHEN UPPER(TRIM(Province)) IN ('NW','NORTH WEST','NORTHWEST') THEN 'North West'
    WHEN UPPER(TRIM(Province)) IN ('LP','LIMPOPO') THEN 'Limpopo'
    WHEN UPPER(TRIM(Province)) IN ('MP','MPUMALANGA') THEN 'Mpumalanga'
    ELSE 'Unknown'
  END AS Province,

  Monthly_Income,
  Join_Date,
  Policy_ID,

  -- Policy_Type normalization (FIXED: added Funeral and Auto variants)
  CASE
    WHEN UPPER(TRIM(Policy_Type)) IN ('LIFE','L') THEN 'Life'
    WHEN UPPER(TRIM(Policy_Type)) IN ('MOTOR','CAR','AUTO') THEN 'Motor'
    WHEN UPPER(TRIM(Policy_Type)) IN ('HOME','HOUSE','PROPERTY') THEN 'Home'
    WHEN UPPER(TRIM(Policy_Type)) IN ('HEALTH','MEDICAL') THEN 'Health'
    WHEN UPPER(TRIM(Policy_Type)) IN ('FUNERAL','F') THEN 'Funeral'
    ELSE 'Unknown'
  END AS Policy_Type,

  Premium_Amount,

  -- Policy_Status normalization
  CASE 
    WHEN UPPER(TRIM(Policy_Status)) IN ('ACTIVE','A') THEN 'ACTIVE'
    WHEN UPPER(TRIM(Policy_Status)) IN ('CANCELLED','CANCELED','C') THEN 'CANCELLED'
    WHEN UPPER(TRIM(Policy_Status)) IN ('LAPSED','L') THEN 'LAPSED'
    WHEN UPPER(TRIM(Policy_Status)) IN ('DORMANT','D') THEN 'DORMANT'
    ELSE 'UNKNOWN'
  END AS Policy_Status,

  Claim_ID,
  Claim_Date,
  Claim_Amount,

  -- Claim_Status normalization
  CASE 
    WHEN UPPER(TRIM(Claim_Status)) IN ('APPROVED','A','IN REVIEW') THEN 'APPROVED'
    WHEN UPPER(TRIM(Claim_Status)) IN ('REJECTED','DENIED','R','DECLINED') THEN 'REJECTED'
    WHEN UPPER(TRIM(Claim_Status)) IN ('PENDING','P') THEN 'PENDING'
    ELSE 'UNKNOWN'
  END AS Claim_Status,

  -- Fraud_Flag normalization
  CASE 
    WHEN UPPER(TRIM(Fraud_Flag)) IN ('Y','YES','1','TRUE','FRAUD') THEN 'Fraud'
    WHEN UPPER(TRIM(Fraud_Flag)) IN ('N','NO','0','FALSE','NON-FRAUD') THEN 'Non-Fraud'
    ELSE 'Unknown'
  END AS Fraud_Flag

FROM insurance_data_clean;

SELECT *
FROM insurance_data_clean
```

```sql
--Remove invalid records
CREATE OR REPLACE TABLE insurance_data_clean AS
SELECT *
FROM insurance_data_clean
WHERE 
    -- IDs must not be Unknown
    Customer_ID <> 'Unknown'
    AND Policy_ID <> 'Unknown'
    AND Claim_ID <> 'Unknown'

    -- Age must be realistic
    AND Age BETWEEN 18 AND 100

    -- Gender must be valid
    AND Gender IN ('Male','Female','Unknown')

    -- Province must be one of the 9 SA provinces or Unknown
    AND Province IN (
        'Western Cape','Eastern Cape','KwaZulu-Natal','Free State',
        'Gauteng','North West','Limpopo','Mpumalanga','Unknown'
    )

    -- Monthly_Income, Premium_Amount, Claim_Amount must be non-negative
    AND Monthly_Income >= 0
    AND Premium_Amount >= 0
    AND Claim_Amount >= 0

    -- Dates must not be placeholder
    AND Join_Date <> DATE('1900-01-01')
    AND Claim_Date <> DATE('1900-01-01')

    -- Policy_Status must be valid
    AND Policy_Status IN ('ACTIVE','CANCELLED','LAPSED','UNKNOWN')

    -- Policy_Type must be valid (FIXED: added Funeral)
    AND Policy_Type IN ('Life','Motor','Home','Health','Funeral','Unknown')

    -- Claim_Status must be valid
    AND Claim_Status IN ('APPROVED','REJECTED','PENDING','UNKNOWN')

    -- Fraud_Flag must be valid
    AND Fraud_Flag IN ('Fraud','Non-Fraud','Unknown')
    
```

```sql
--View Final clean insurance table
SELECT *
FROM insurance_data_clean
```
